# <mark></mark>Worldwide Earthquake Events API - Bronze Layer Processing

```python
import requests
import json

# Construire l'URL de l'API avec les dates de début et de fin fournies par Data Factory,
# formatée pour une sortie geojson.
url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={start_date}&endtime={end_date}"

# Envoyer la requête GET pour récupérer les données
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Récupérer la réponse au format JSON
    data = response.json()
    data = data['features']
    
    # Spécifier le nom du fichier 
    file_path = f'/lakehouse/default/Files/{start_date}_earthquake_data.json'
    
    # Ouvrir le fichier en mode écriture ('w') et sauvegarder les données JSON
    with open(file_path, 'w') as file:
        # La méthode `json.dump` sérialise `data` en JSON et l'écrit dans `file`
        # `indent=4` rend le fichier lisible en ajoutant des espaces
        json.dump(data, file, indent=4)
        
    print(f"Les données ont été sauvegardées avec succès dans {file_path}")
else:
    print("Échec de la récupération des données. Code de statut :", response.status_code)
```
